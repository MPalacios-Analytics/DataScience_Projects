```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('500 richest people 2021.csv')
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
df.describe()
```

```python
df[df['Total Net Worth'].str.contains('[A-Za-z]', regex=True)]
```

```python
df[df['Total Net Worth'] == 'family']
df = df[df['Total Net Worth'] != 'family']
```

```python
df['Total Net Worth'] = df['Total Net Worth'].str.replace('$','', regex=False)
df['Total Net Worth'] = df['Total Net Worth'].str.replace('B','', regex=False)
df['Total Net Worth'] = df['Total Net Worth'].astype(float)
```

```python
df.info()
```

## Visualization 1

```python
fig, ax = plt.subplots(figsize=(8,5))

ax.hist(df['Total Net Worth'], bins=20)
ax.set_title("Distribution of Net Worth Among the 500 Richest People")
ax.set_xlabel("Net Worth (Billions USD)")
ax.set_ylabel("Number of Individuals")
plt.show()
```

## Visualization 2

```python
top_countries = df['Country'].value_counts().head(10).sort_values()
fig, ax = plt.subplots(figsize=(8,5))
ax.barh(top_countries.index, top_countries.values)
ax.set_title("Top 10 Countries with the Most Billionaires")
ax.set_xlabel("Number of Individuals")
ax.set_ylabel("Country")

plt.show()
```

## Visualization 3

```python
top_industries = df['Industry'].value_counts().head(10)
top_industries = top_industries.sort_values()
fig, ax = plt.subplots(figsize=(8,5))
ax.barh(top_industries.index, top_industries.values)
ax.set_title("Top Industries Among the 500 Richest Individuals")
ax.set_xlabel("Number of Individuals")
ax.set_ylabel("Industry")


plt.show()
```
